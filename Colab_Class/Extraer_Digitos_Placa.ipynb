```python
%matplotlib inline
from matplotlib import pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
placaRGB = cv2.imread('placa.png')
```

```python
placaRGB.shape
```

```python
cv2_imshow(placaRGB)
```

```python
placaBW= cv2.cvtColor(placaRGB, cv2.COLOR_RGB2GRAY) 
```

```python
cv2_imshow(placaBW)
```

```python
placa_bw_neg = 255 - placaBW
```

```python
cv2_imshow(placa_bw_neg)
```

```python
proy_horizontal = np.sum(placa_bw_neg,axis=1)
plt.plot(proy_horizontal)
```

```python
linea_1 = placa_bw_neg[0:240,:]
cv2_imshow(linea_1)
```

```python
proy_vertical = np.sum(linea_1,axis=0)
plt.plot(proy_vertical)
```
